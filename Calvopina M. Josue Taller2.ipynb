## Graficar la funcion

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definir la función matemática


def f(x):
    return x**3 - 3*x**2 + x - 1


# 2. Generar un conjunto de valores para el eje X
x = np.linspace(-2, 4, 400)

# 3. Calcular los valores correspondientes en el eje Y
y = f(x)

# 4. Configurar y dibujar la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^3 - 3x^2 + x - 1$', color='blue')

# Dibujar los ejes X e Y que cruzan por el origen (0,0)
plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)

# --- NUEVO: Añadir números (marcas) explícitos en el eje X ---
# np.arange(-2, 5, 1) crea una lista desde -2 hasta 4, de 1 en 1.
plt.xticks(np.arange(-2, 5, 1))

# Añadir etiquetas y título
plt.title('Gráfica de la función $f(x) = x^3 - 3x^2 + x - 1$')
plt.xlabel('x')
plt.ylabel('f(x)')

# Añadir cuadrícula y la leyenda
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Mostrar la gráfica
plt.show()
```

## Metodo del profesor

```python
xs = list(range(-10, 30))
```

```python
def f(x):
    _y = x**3 - 3 * x**2 + x - 1
    return _y
```

```python
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt

plt.plot(xs, ys)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Gráfica de ys vs xs")
plt.grid(True)
plt.show()
```

## Metodo de newton aplicado

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize


def f(x):
    return x**3 - 3*x**2 + x - 1


root = optimize.newton(f, 100)
print(root)

root2 = optimize.newton(f, 1)
print(root2)

root2 = optimize.newton(f, -0.75)
print(root2)
```
